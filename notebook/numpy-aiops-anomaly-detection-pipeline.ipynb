```python
# AIOps Mini project - numpy-aiops-anomaly detections pipeline
# A complete end to end AIOps monitoring pipeline in pure NumPy. It does exactly what a real monitoring system does — generate metrics, detect anomalies using three methods, correlate alerts across metrics, and produce an incident report.
```

```python
# Generating metrics
import numpy as np
rng = np.random.default_rng(seed=42)
n = 180 # 3 hours one reading per minute

server_names = ['web-01', 'app-01', 'db-01']

cpu = np.column_stack([
    np.clip(rng.normal(60,5,n),0,100), #web-01 healthy
    np.clip(rng.normal(72,6,n),0,100), #app-01 running hot
    np.clip(rng.normal(45,4,n),0,100), #db-01 low usage
])

memory = np.column_stack([
    np.clip(rng.normal(65,8,n),0,100),
    np.clip(rng.normal(78,10,n),0,100),
    np.clip(rng.normal(55,6,n),0,100),
])

latency = np.column_stack([
    np.clip(rng.normal(18,3,n),0,500),
    np.clip(rng.normal(25,4,n),0,500),
    np.clip(rng.normal(12,2,n),0,500),
])

print(f"servers: {server_names}")
print(f"cpu shape: {cpu.shape}")
print("cpu metrics\n", cpu[:10].round(1)) # sample - first 10 elements
print(f"memory shape: {memory.shape}")
print("memory metrics\n", memory[:10].round(1)) # sample -frist 10 elements
print(f"latency shape: {latency.shape}")
print("latency metrics:", latency[:10].round(1))
```

```python
# baseline stastics per server
print("\n-------------baseline statistics--------------")

for i, name in enumerate(server_names):
    print(f"\n{name}")
    
    print(f"  cpu mean     : {cpu[:,i].mean():.1f}%   "
          f"std: {cpu[:,i].std():.1f}%   "
          f"p95: {np.percentile(cpu[:,i],95):.1f}%   ")

    print(f"  memory mean  : {memory[:,i].mean():.1f}%   "
          f"std: {memory[:,i].std():.1f}%   "
          f"p95: {np.percentile(cpu[:,i],95):.1f}%   " )

    print(f"  latency mean : {latency[:,i].mean():.1f}%   "
          f"std: {latency[:,i].mean():.1f}%   "
          f"p95: {np.percentile(latency[:,i],95):.1f}ms ")
```

```python
# inject incidents and detect anomalies
inc_a = np.arange(45, 48)
cpu[inc_a,0] = [91, 94, 89]
memory[inc_a,0] = [92, 95, 90]
latency[inc_a,0] = [85, 110, 92] 

inc_b = np.arange(120, 123)
cpu[inc_b,1] = [95, 98, 93]
memory[inc_b, 1] = [97, 99, 95]
latency[inc_b] = [120, 145, 118]

def detect_anomalies(data, static_threshold=85):
    static = set(np.where(data > static_threshold)[0])

    z = (data - data.mean()) / data.std()
    zscore = set(np.where(np.abs(z) > 2)[0])

    q25, q75 = np.percentile(data, [25, 75])
    iqr = q75 - q25
    iqr_set = set(np.where(
        (data < q25 - 1.5*iqr) | (data > q75 + 1.5*iqr)
    )[0])
    consensus = (static & zscore) | (static & iqr_set) | (zscore & iqr_set)
    return consensus
print("\n ------- anomaly detection -------")
for i, name in enumerate(server_names):
    cpu_alerts = detect_anomalies(cpu[:,i])
    memory_alerts = detect_anomalies(memory[:,i])
    latency_alerts = detect_anomalies(latency[:,i], static_threshold=50)

    print(f"\n {name} ")
    print(f"cpu alerts   : {sorted(cpu_alerts)}")
    print(f"memory alerts : {sorted(memory_alerts)}")
    print(f"latency alerts : {sorted(latency_alerts)} ")
```

```python
# Correlate alerts across metrics
print("\n--------- Correlated Incidents ---------")

for i, name in enumerate(server_names):
    cpu_a = detect_anomalies(cpu[:, i])
    mem_a = detect_anomalies(memory[:, i])
    lat_a = detect_anomalies(latency[:, i], static_threshold=50)

    # intersection of all metrics
    incident_idx = cpu_a & mem_a & lat_a

    if len(incident_idx) > 0:
        print(f"\n{name} - confirmed incident at minutes: {sorted(incident_idx)}")

        for idx in sorted(incident_idx):
            print(
                f" minute {idx:3d} | "
                f"cpu: {cpu[idx, i]:.1f}% | "
                f"mem: {memory[idx, i]:.1f}% | "
                f"lat: {latency[idx, i]:.1f} ms"
            )
    else:
        print(f"\n{name} - no confirmed incidents")
```

```python
# summary dashboard
print("\n --- summanry dashboard ----")
print(f"server     cpu mean     memory mean "
      f"    lat p95      incidents " )
print("-"*62)

for i,name in enumerate (server_names):
    cpu_a = detect_anomalies(cpu[:,i])
    mem_a = detect_anomalies(memory[:,i])
    list_a = detect_anomalies(latency[:,i], static_threshold=50)
    incidents = len(cpu_a & mem_a & lat_a)

    print(f"{name}       "
          f"{cpu[:,i].mean().round(1)}       "
          f"{memory[:,i].mean().round(1)}              "
          f"{np.percentile(latency[:,i],95).round(1)}             "
          f"{incidents}" )
```
